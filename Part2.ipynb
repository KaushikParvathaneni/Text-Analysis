```python
positive = list(positive)
negative = list(negative)
```

```python
maximum_list = list(new['FILTERED'])
minimum_list = list(new['minimal_preprocessing'])
```

```python
def pol_cal(list_data):    
    positive_pol = []
    negative_pol = []
    for sentence in list_data:
        count_positive = 0
        count_negative = 0
        sent_len = 0
        for word in sentence.split():
            sent_len = len(sentence.split())
            if word in positive:
                count_positive +=1
            elif word in negative:
                count_negative +=1
        positive_pol.append(count_positive/sent_len)
        negative_pol.append(count_negative/sent_len)
    return positive_pol, negative_pol
            
```

```python
positive_pol_max = pol_cal(maximum_list)[0]
negative_pol_max = pol_cal(maximum_list)[1]
```

```python
positive_pol_min = pol_cal(minimum_list)[0]
negative_pol_min = pol_cal(minimum_list)[1]
```

#### -What are the average polarity scores?

```python
# What are the average polarity scores?
print('avg pos polarity of maximum filtered data:',np.array(positive_pol_max).mean())
print('avg neg polarity of maximum filtered data:',np.array(negative_pol_max).mean())
print('avg pos polarity of minimum filtered data:',np.array(positive_pol_min).mean())
print('avg neg polarity of minimum filtered data:',np.array(negative_pol_min).mean())

average_max = (np.array(positive_pol_max) + np.array(negative_pol_max)).sum()/len(np.array(positive_pol_max) + np.array(negative_pol_max))
print('Average polarity of all the max filtered data', average_max)

average_min = (np.array(positive_pol_min) + np.array(negative_pol_min)).sum()/len(np.array(positive_pol_min) + np.array(negative_pol_min))
print('Average polarity of all the min filtered data', average_min)
```

#### -	What are the average polarity scores across different kinds of stances?

```python
# What are the average polarity scores across different kinds of stances? 
def fil_new(new,unq_col_values, col_name):
    '''This function is used to filter the dataframes 
    Inputs:dataframe name (df), list of unique column names in that column (unique_col_values), the Column name it self(col_name)
    returns dataframes of each unique column name that was filtered
    '''
    new_frame = []
    for i in unq_col_values:
        new_frame.append(new[new[col_name] == i]) 
    return new_frame
```

```python
st = new['Stance'].unique()
print(st)
```

```python
new['positive_pol_max'] = positive_pol_max
new['negative_pol_max'] = negative_pol_max
new['positive_pol_min'] = positive_pol_min
new['negative_pol_min'] = negative_pol_min
```

```python
new['maximum_sum_of_pol'] = np.array(positive_pol_max) + np.array(negative_pol_max)
new['minimum_sum_of_pol'] = np.array(positive_pol_min) + np.array(negative_pol_min)
```

```python
for i in range(0,len(st)):
    st_new = fil_new(new, st, 'Stance' )[i]
    temp_sum = np.array(fil_new(new, st, 'Stance' )[i]['maximum_sum_of_pol'])
    print(f'{st[i]}:')
    print(f'Average Positive polarity score for \'{st[i]}\' Stance  for max processing:' , st_new['positive_pol_max'].mean()) 
    print(f'Average Negative polarity score for \'{st[i]}\' Stance  for max processing:' , st_new['negative_pol_max'].mean()) 
    print(f'Average Polarity Score for \'{st[i]}\' Stance for max processing:' ,temp_sum.sum()/len(temp_sum), '\n')
```
