```python
list_max = list(df['filtered_max'])
list_min = list(df['filtered_min'])
```

```python
def polarity_calc(list_data): 
    '''This function calculates the polarity of a list of sentences and returns the list of positive polarity and 
    negative polarity values of  each sentences
    Note: pass in only list data
    polarity_calc[0] -> positive polarity values
    polarity_calc[1] -> negative polarity values
    
    '''   
    pos_polarity_sent = []
    neg_polarity_sent = []
    for sentence in list_data:
        count_pos = 0
        count_neg = 0
        sent_length = 0
        for word in sentence.split():
            sent_length = len(sentence.split())
            if word in pos_words:
                count_pos +=1
            elif word in neg_words:
                count_neg +=1
        pos_polarity_sent.append(count_pos/sent_length)
        neg_polarity_sent.append(count_neg/sent_length)
    return pos_polarity_sent, neg_polarity_sent
            
```

```python
pos_polarity_max_data = polarity_calc(list_max)[0]
neg_polarity_max_data = polarity_calc(list_max)[1]
```

```python
pos_polarity_min_data = polarity_calc(list_min)[0]
neg_polarity_min_data = polarity_calc(list_min)[1]
```

```python
# #-	What are the average polarity scores?
print('Average positive polarity of max filtered data:',np.array(pos_polarity_max_data).mean())
print('Average negative polarity of max filtered data:',np.array(neg_polarity_max_data).mean())
print('Average positive polarity of min filtered data:',np.array(pos_polarity_min_data).mean())
print('Average negative polarity of min filtered data:',np.array(pos_polarity_min_data).mean())

avg_max = (np.array(pos_polarity_max_data) + np.array(neg_polarity_max_data)).sum()/len(np.array(pos_polarity_max_data) + np.array(neg_polarity_max_data))
print('Average polarity of all the max filtered data', avg_max)

avg_min = (np.array(pos_polarity_min_data) + np.array(neg_polarity_min_data)).sum()/len(np.array(pos_polarity_min_data) + np.array(neg_polarity_min_data))
print('Average polarity of all the min filtered data', avg_min)

```

```python
# -	What are the average polarity scores across different kinds of stances?



def filter_data_frame(df,unique_col_vals, col_name):
    '''This function is used to filter the dataframes 
    Inputs:dataframe name (df), list of unique column names in that column (unique_col_values), the Column name it self(col_name)
    returns dataframes of each unique column name that was filtered
    '''
    data_frames = []
    for i in unique_col_vals:
        data_frames.append(df[df[col_name] == i]) 
    return data_frames
```
